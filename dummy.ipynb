```python
import tensorflow as tf
```

```python
sess = tf.InteractiveSession()
```

```python
d = tf.Variable(tf.random.normal(shape=[10,91]))
```

```python
pred = tf.Variable(tf.random.normal(shape=[10,91]))
```

```python
sess.run(tf.global_variables_initializer())
```

```python
mask = d > 0
```

```python
ne = tf.boolean_mask(d,mask)
```

```python
pred_new = tf.boolean_mask(pred,mask)
```

```python
d.shape, mask.shape
```

```python
new = tf.dtypes.cast(mask,'float')
```

```python
sess.run(tf.math.multiply(d,new))
```

```python
loss = tf.losses.softmax_cross_entropy(logits=pred_new,labels=ne)
```

```python
sess.run(loss)
```
